```python
import numpy as np
import matplotlib.pyplot as plt
import math
from math import factorial
from sympy import *
```

```python
#Ex2.1
N=10000
seed=100
np.random.seed(seed)
x=np.random.random(N)
seed=1000
np.random.seed(seed)
y=np.random.random(N)
i=0
k=0
r=0.5
for c in range(N):
    if (x[c]-0.5)**2+(y[c]-0.5)**2 <= 0.25:
        i=i+1
    else:
        k=k+1
A=i/N
print(A)
plt.scatter(x,y,s=0.01)
plt.show()

r=0.5
A=math.pi*(r**2)
print(A)
```

```python
#Ex2.2
error=[]
n=[]
Y=[]
for i in range(1,8):
    N=10**i
    seed=100
    np.random.seed(seed)
    x=np.random.random(N)
    seed=1000
    np.random.seed(seed)
    y=np.random.random(N)
    i=0
    k=0
    r=0.5
    for c in range(N):
        if (x[c]-0.5)**2+(y[c]-0.5)**2 <= 0.25:
            i=i+1
        else:
            k=k+1
    A=i/N
    #print(A)
    #plt.show()
    r=0.5
    a=math.pi*(r**2)
    #print(A)
    error.append(np.log10(abs((A-a)/a)))
    n.append(np.log10(N))
    print(N,1/np.sqrt(N))
    Y.append(1/np.sqrt(N))
    
```

```python
plt.plot(n,error,'r',label='compute error')
plt.plot(n,np.log10(Y),'--',label='expect scaling')
z=np.polyfit(n,error,1)
p1=np.poly1d(z)
plt.xlabel('log10 N')
plt.ylabel('log10 error')
plt.legend()
plt.title('log10(error) V.S. log10 N')
plt.savefig('plot2.2.png')
plt.show()
```

```python
#Ex2.3
error=[]
n=[]
Y=[]
for i in range(1,8):
    N=10**i
    seed=100
    np.random.seed(seed)
    x=np.random.random(N)*2
    seed=1000
    np.random.seed(seed)
    y=np.random.random(N)*4
    i=0
    k=0
    r=0.5
    for c in range(N):
        if y[c] <= x[c]**2:
            i=i+1
        else:
            k=k+1
    A=8*i/N
    #print(A)
    #plt.show()
    r=0.5
    a=(1/3)*(2**3)
    #print(A)
    error.append(np.log10(abs((A-a)/a)))
    n.append(np.log10(N))
    print(N,error)
    Y.append(1/np.sqrt(N))
```

```python
plt.plot(n,error,'r',label='compute error')
plt.plot(n,np.log10(Y),'--',label='expect scaling')
z=np.polyfit(n,error,1)
p1=np.poly1d(z)
plt.xlabel('log10 N')
plt.ylabel('log10 error')
plt.legend()
plt.title('log10(error) V.S. log10 N')
plt.savefig('plot2.3.png')
plt.show()
```

```python
#EX6.2
h=1
H=[]
error=[]
for i in range(100):
    h=0.9*h
    x=0.1
    def y(x):
        z=math.cos(x)
        return z
    A=(y(x+h/2)-y(x-h/2))/h 
    a=-math.sin(x)
    H.append(math.log(h, 10))
    error.append(math.log(abs((A-a)/a),10))

scatter=plt.scatter(H,error)      
z=np.polyfit(H,error,1)
p1=np.poly1d(z)
ax = scatter.axes
ax.invert_xaxis()
#ax.invert_yaxis()
plt.xlabel('log10 h')
plt.ylabel('log10 error')
print(p1)
plt.title('log10(error) V.S. log10 h')
plt.savefig('plot6.2.png')
plt.show()
```
