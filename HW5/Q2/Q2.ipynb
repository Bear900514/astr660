```python
#Q2.3 LF
a=np.loadtxt('C:/Users/Bear/OneDrive/文件/程式/HW5/Q2/2.3/50/advection_LF00080.d')
b=np.loadtxt('C:/Users/Bear/OneDrive/文件/程式/HW5/Q2/2.3/500/advection_LF00870.d')
c=np.loadtxt('C:/Users/Bear/OneDrive/文件/程式/HW5/Q2/2.3/5000/advection_LF08750.d')
N=[50,500,5000]
name1=80
name2=870
name3=8750
n1=50
n2=500
n3=5000
dx1=1/n1
dx2=1/n2
dx3=1/n3
cfl=0.4
dt1=dx1*cfl
dt2=dx2*cfl
dt3=dx3*cfl
time1=dt1*(name1)
time2=dt2*(name2)
time3=dt3*(name3)
x1=a[:,1]
U1=a[:,2]
u1=[0]*50
error1=0
all_error=[]
for i in range(50):
    u1[i]=math.exp(-1000*(x1[i]-time1*1-0.1)**2)
    error1=error1+abs(U1[i]-u1[i])
error1=error1/50
all_error.append(error1)

x2=b[:,1]
U2=b[:,2]
u2=[0]*500
error2=0
for i in range(500):
    u2[i]=math.exp(-1000*(x2[i]-time2*1-0.1)**2)
    error2=error2+abs(U2[i]-u2[i])
error2=error2/500
all_error.append(error2)

x3=c[:,1]
U3=c[:,2]
u3=[0]*5000
error3=0
for i in range(5000):
    u3[i]=math.exp(-1000*(x3[i]-time3*1-0.1)**2)
    error3=error3+abs(U3[i]-u3[i])
error3=error3/5000
all_error.append(error3)

#print(all_error)
plt.plot(N,all_error)
plt.loglog()
plt.xlabel('N')
plt.ylabel('error')
#plt.rcParams['figure.figsize'] = [15,5]
plt.title('LF error vs N')
plt.savefig('Q2.3LF.png')
plt.show()
```
