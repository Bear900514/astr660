```python
A=np.loadtxt('C:/Users/Bear/OneDrive/文件/程式/HW4/euler/euler-0.001.txt')
B=np.loadtxt('C:/Users/Bear/OneDrive/文件/程式/HW4/euler/euler-0.01.txt')
C=np.loadtxt('C:/Users/Bear/OneDrive/文件/程式/HW4/euler/euler-0.1.txt')
D=np.loadtxt('C:/Users/Bear/OneDrive/文件/程式/HW4/euler/euler-1.0.txt')

t=0
n=len(A[:,0])
theta=math.pi/3
vx=30*math.cos(theta)
vy=30*math.sin(theta)
x=vx*t
y=vy*t-(1/2)*9.8*t**2
error=[]
sum1=0
for i in range(n):
    t=A[i,0]
    y=vy*t-(1/2)*9.8*t**2
    sum1=sum1+abs(A[i,2]-y)
error.append(sum1/n)

n=len(B[:,0])

sum2=0
for i in range(n):
    t=B[i,0]
    y=vy*t-(1/2)*9.8*t**2
    sum2=sum2+abs(B[i,2]-y)
error.append(sum2/n)

n=len(C[:,0])
sum3=0
for i in range(n):
    t=C[i,0]
    y=vy*t-(1/2)*9.8*t**2
    sum3=sum3+abs(C[i,2]-y)
error.append(sum3/n)

n=len(D[:,0])
sum4=0
for i in range(n):
    t=D[i,0]
    y=vy*t-(1/2)*9.8*t**2
    sum4=sum4+abs(D[i,2]-y)
error.append(sum4/n)
print(error)
dt=[0.001,0.01,0.1,1.0]
plt.scatter(dt,error)
plt.loglog()
plt.xlabel('log(dt)')
plt.ylabel('log(error)')
#plt.rcParams['figure.figsize'] = [15,5]
plt.title('dt-error with euler method')
plt.savefig('Q1.2euler.png')
plt.show()
```
