```python
import numpy as np
import matplotlib.pyplot as plt
import math
from math import factorial
```

```python
#Pro5.1
x=1
value=[]
all_error=[]
b=range(1,101)
true=math.exp(-x)
for j in range(1,101):
    a=0
    for i in range(j+1):
        k=(-x)**(i)/factorial(i)
        a=a+k
    error=(a-true)/true
    A=math.log10(abs(error))
    all_error.append(A)
    value.append(a)
    
plt.subplot(121)
plt.plot(b,value)
plt.scatter(b,value) 
plt.title(r'summation for x =1.0')
plt.ylabel('exponential of value')
plt.xlabel('N')

plt.subplot(122)
plt.plot(b,all_error)
plt.scatter(b,all_error) 
plt.title(r'relative error for x =1.0')
plt.ylabel('relative error(with log10 basic)')
plt.xlabel('N')
plt.rcParams['figure.figsize'] = [15,5]

plt.savefig('plot1')
print(all_error)
```
